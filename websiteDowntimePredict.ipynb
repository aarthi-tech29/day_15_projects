### Project 1: Website Downtime Prediction 
#### Objective: Use a Poisson distribution to model the number of server crashes per month and calculate the probability of having a certain number of crashes. 

##### Step 1: Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

#####  Step 2: Set Parameters

```python
lam = 3  # λ = average number of crashes per month

# the server crashes 3 times per month on average
```

##### Step 3: Calculate probabilities

```python
# Probability of exactly 2 crashes
p_2 = poisson.pmf(2, lam)

# Probability of at most 2 crashes
p_upto_2 = poisson.cdf(2, lam)

print(f"Probability of exactly 2 crashes: {p_2:.3f}")
print(f"Probability of at most 2 crashes: {p_upto_2:.3f}")

# Probability Mass Function (PMF) - It calculates the probability of exactly 2 crashes in a month
# Cumulative Distribution Function (CDF) - It finds the probability of having 0, 1, or 2 crashes (up to 2)
# poisson.pmf(k, lam) → Probability of exactly k crashes.
# poisson.cdf(k, lam) → Probability of at most k crashes.
```

##### Step 5: Visualize the distribution

```python
x = np.arange(0, 11)  # possible crash counts
y = poisson.pmf(x, lam)

plt.bar(x, y, color='skyblue', edgecolor='black')
plt.title("Poisson Distribution: Website Server Crashes per Month")
plt.xlabel("Number of Crashes")
plt.ylabel("Probability (PMF)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

# x → number of crashes (0 to 10)
# y → probability of each crash count - This calculates the PMF for each crash count in x.y stores the probability of 0, 1, 2, …, 10 crashes.
#  eg. for y stores =  y = [0.0498, 0.1494, 0.2240,....]
# The highest bar near 3 (since λ=3).
# Smaller bars as you move away — because more or fewer crashes are less likely.
# It’s likely that the site will crash 2–3 times per month.
```
