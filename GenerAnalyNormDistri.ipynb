### Project 2: Generating and Analyzing a Normal Distribution
#### Objective: Generate a random dataset that follows a normal distribution and verify its properties. 

##### Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Step 2: Generate Random Normal Data

```python
mu = 170     # mean
sigma = 5    # standard deviation
data = np.random.normal(mu, sigma, 1000)  # 1000 data points

# mu (μ) = mean → the center of the bell curve
# sigma (σ) = standard deviation → how spread out the data is
# np.random.normal(μ, σ, n) generates random data following the Normal Distribution
```

##### Step 3: Statistical Analysis

```python
calculated_mean = np.mean(data)
calculated_std = np.std(data)

# np.mean(data) → finds the average of all numbers
# np.std(data) → finds the standard deviation
```

##### Step 4: Check the Empirical Rule

```python
within_1_sigma = np.sum((data > mu - sigma) & (data < mu + sigma))
within_2_sigma = np.sum((data > mu - 2*sigma) & (data < mu + 2*sigma))
within_3_sigma = np.sum((data > mu - 3*sigma) & (data < mu + 3*sigma))

# how many data points fall:
# Within 1σ (68%) → between 165 and 175
# Within 2σ (95%) → between 160 and 180
# Within 3σ (99.7%) → between 155 and 185
# np.sum() counts how many values meet each condition
```

##### Step 5: Convert Counts to Percentages

```python
n = len(data)
p1 = within_1_sigma / n * 100
p2 = within_2_sigma / n * 100
p3 = within_3_sigma / n * 100

print(p1)
print(p2)
print(p3)

# n = len(data) → total number of data points (1000)
# p = percentage of data within the σ ranges
```

##### Step 6: Visualization (Histogram)

```python
plt.figure(figsize=(8, 5))
sns.histplot(data, bins=30, kde=True, color='skyblue') # bins=30 → divides data into 30 intervals
plt.title('Normal Distribution (μ=170, σ=5)') # kde-Kernel Density Estimate curve
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

##### Step 7: Output Results


```python
print(f"Calculated Mean: {calculated_mean:.2f}")
print(f"Calculated Standard Deviation: {calculated_std:.2f}")
print(f"Within 1σ: {p1:.2f}%")
print(f"Within 2σ: {p2:.2f}%")
print(f"Within 3σ: {p3:.2f}%")
```
