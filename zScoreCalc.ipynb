### Project 2: Z-Score Calculator 
#### Objective: Write a function to calculate the Z-score for a given value, mean, and standard deviation. 

##### Step 1: Import NumPy

```python
import numpy as np
```

##### Step 2: Dataset

```python
marks = [72, 65, 90, 58, 81, 75, 69, 95, 88, 20]
```

##### Step 3: Find Mean (μ) and Standard Deviation (σ)

```python
mu = np.mean(marks)
sigma = np.std(marks)
```

##### Step 4: Define the Z-score formula

```python
def z_score(x, mu, sigma):
    return (x - mu) / sigma

#  Formula: Z = (X−μ​) / σ
# It measures how far a value x is from the mean, in units of standard deviation
```

##### Step 5: Calculate Z-score for each student

```python
z_scores = [z_score(x, mu, sigma) for x in marks]
```

##### Step 6: Find Outliers

```python
outliers = [(marks[i], z_scores[i]) for i in range(len(marks)) if abs(z_scores[i]) > 2]


#  If abs(z_score) (absolute value) greater than 2
#  If yes → that value is too far from the mean → outlier
#  abs() -> it gives you how far a number is from zero, without caring if it’s positive or negative.
#  abs() — so we check both high and low extreme values at once.
# "1" - close to average
# "2" means a little suspicious - abs(z_scores[i]) > 2
# "3" means definitely weird
```

##### Step 7: Print Results

```python
print("Marks:", marks)
print(f"Mean (μ): {mu:.2f}")
print(f"Standard Deviation (σ): {sigma:.2f}")
print("Z-scores:", [round(float(z), 2) for z in z_scores])
```

##### Step 8: Print Outliers

```python
if outliers:
    print("\n Outliers found:")
    for value, z in outliers:
        print(f"Mark: {value}, Z-score: {z:.2f}")
else:
    print("\n No outliers found.")
```
