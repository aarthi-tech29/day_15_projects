### Project 1: Simulating and Visualizing a Poisson Distribution
#### Objective: Simulate the number of customers arriving at a coffee shop and visualize the distribution of arrivals. 

##### Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Step 2: Define Parameter

```python
lam = 5  # lambda - average number of customers per minute,5 customers arrive per minute.
intervals = 1000  # number of time intervals to simulate, each interval will have a random number of customers
```

##### Step 3: Simulate Poisson distribution

```python
arrivals = np.random.poisson(lam, intervals)
```

##### Step 4: Visualize the results

```python
plt.figure(figsize=(8, 5))
plt.hist(arrivals, bins=range(0, 16), align='left', rwidth=0.8)
plt.title("Poisson Distribution of Customer Arrivals at a Coffee Shop")
plt.xlabel("Number of Customers Arriving per Minute")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# bins=range(0,16) → makes bars for customer counts 0, 1, 2, … up to 15
# rwidth=0.8 → makes a small gap between bars
#  grid - Adds horizontal dashed lines to help read the chart
# frequency - y axis - how many times that number of customers arrived according to the 1000 intervals —that’s called the frequency. 
```

##### Step 5: Analysis

```python
mean_arrivals = np.mean(arrivals)
print(f"Average number of arrivals (simulated): {mean_arrivals:.2f}")
```
